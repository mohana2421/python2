### ***Matplot Library***

```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5,6,7,8]
y=[2,5,7,8,9,5,7,8]
plt.plot(x,y)
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5,6,7,8]
y=[2,5,7,8,9,5,7,8]
plt.bar(x,y)
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5,6,7,8]
###y=[2,5,7,8,9,5,7,8]
plt.pie(x)
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5,6,7,8]
y=[2,5,7,8,9,5,7,8]
plt.scatter(x,y)
plt.show()
```

```python
### this is for line plot 
from matplotlib import pyplot as plt
x=[1,2,3,4,5,6,7,8]
y=[2,5,7,8,9,5,7,8]
plt.bar(x,y)
plt.xlabel("speed")
plt.ylabel("distance")
plt.show()
```
